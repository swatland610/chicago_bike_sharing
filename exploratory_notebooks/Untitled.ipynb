```python
import pandas as pd 
pd.set_option("display.max_columns", 101)
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns

# import data
divvy_2019 = pd.read_csv('../cleaned_data/divvy_2019_trips.csv')
divvy_2020 = pd.read_csv('../cleaned_data/divvy_2020_trips.csv')
```

```python
divvy_2019.info()
```

```python
# GroupBy Objects
# Chicago Area Counts
divvy_2019_trips_by_chicago_area = divvy_2019.groupby('from_chicago_area').agg({'trip_id':'size', 'from_station_id':'nunique'})
divvy_2020_trips_by_chicago_area = divvy_2020.groupby('from_chicago_area').agg({'ride_id':'size', 'start_station_id':'nunique'})
```

```python
# Pivot Tables
divvy_2019_from_to_community_area = pd.pivot_table(divvy_2019, 
                                                   index=['from_chicago_area', 'from_community_name'], 
                                                   columns=['to_chicago_area', 'to_community_name'],
                                                  values=['trip_id'],
                                                  aggfunc='size',
                                                  fill_value=0)

divvy_2020_from_to_community_area = pd.pivot_table(divvy_2020, 
                                                   index=['from_chicago_area', 'from_community_name'], 
                                                   columns=['to_chicago_area', 'to_community_name'],
                                                  values=['ride_id'],
                                                  aggfunc='size',
                                                  fill_value=0)
```
